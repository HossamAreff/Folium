```python
world_map = folium.Map()
world_map.save("WorldMap.html")
world_map
```

```python
egypt=folium.Map(location=[30.033333, 31.233334],zoom_start=7)
folium.Marker([30.033333, 31.233334], popup= 'Egypt').add_to(egypt)
folium.CircleMarker(location=[],radius=50,popup='Laurelhurst Park',
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(egypt)
```

```python
folium.TileLayer('openstreetmap').add_to(egypt)
folium.TileLayer('cartodbpositron').add_to(egypt)
```

```python
folium.LayerControl().add_to(egypt)
```
